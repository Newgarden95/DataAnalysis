# 빌보드차트 데이터를 통해 2020~2021년 사이 
# 가장 많이 1위를 차지한 가수 Top9 확인

```python
#분석 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#빌보드 차트 csv파일 로드
billboard = pd.read_csv('charts.csv')
billboard
```

```python
# 날짜를 인덱스로 변경
billboard.set_index('date', inplace=True)
billboard
```

```python
#빌보드 데이터 중 2020~2021년 까지의 데이터 추출
billboard_data = billboard.loc['2020-04-01': , : ]
billboard_data
```

```python
# 1위 데이터만 추출
billboard_data_rank = billboard_data[billboard_data['rank']==1]
billboard_data_rank
```

```python
# 불필요한 데이터 제거 하기
billboard_data_rank2 = billboard_data_rank.drop(columns=['song','last-week','peak-rank','weeks-on-board'])
billboard_data_rank2
```

```python
billboard_data_rank2.shape
billboard_data_rank2['artist'].value_counts() == 1
billboard_data_rank2['artist'].value_counts()[billboard_data_rank2['artist']
                                              .value_counts() == 1]
billboard_data_rank2
```

```python
# Top 9데이터 추출(단순하게 head(9)를 통해 확인)
billboard_top9 = billboard_data_rank['artist'].value_counts().head(9)
billboard_top9
```

```python
# 추출데이터 시각화
plt.figure(figsize=(16,8))
ex = [0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04, 0.04]
billboard_top9.plot.pie(explode = ex,autopct = '%d%%')
plt.title('Billboard Rank Count 2020 ~ 2021')
```

```python
plt.figure(figsize=(16,8))
billboard_top9.plot.bar()
```
