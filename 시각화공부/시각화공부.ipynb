# 다변수 시각화

* X : 연속 / Y : 연속
    - 산점도 / 회귀선 / 상관분석

```python
#기본적으로 시각화 이전 수행할 것
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

mpl.rc('font', family = 'Malgun Gothic')
```

```python
df1 = pd.read_csv('Data03.csv')
df1
```

```python
#연속형 변수!!!!
sns.scatterplot(data = df1, x = 'age', y = 'Mileage') #scatterplot : 산점도
```

```python
#오버레이 기능
sns.scatterplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract') #범주형 항목에 대해 오버레이
```

```python
sns.scatterplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract', size = 'Term')
```

```python
#회귀식~> 선형 회귀선 그리기
sns.lmplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract') # 회귀선 : 점들을 대표하는 선
```

```python
sns.jointplot(data = df1, x = 'age' , y = 'Mileage')
```

```python
#pairplot
sns.pairplot(df1) # 데이터 프레임형태를 넣어주면됨, 산점도 + 분포를 그려줌(모든 숫자연속형 데이터들만)
```

```python
#전체 연속형 데이터를 다 넣으면 시간 많이 걸림 따라서, 내가 원하는 것만 출력할 때 아래와 같이
plt.figure(figsize = [20,20])
sns.pairplot(df1[['age','Mileage','amount','product']], hue = 'product') #product는 범주형인데 왜 넣냐? 오버레이하려고
```

* X : 범주형 / Y : 연속형
  - 학년(X)에 따른 키의 평균(Y) ~> Barplot / Boxplot

```python
plt.figure(figsize = [70, 20])
sns.barplot(data = df1, x = 'channel' , y = 'amount') #디폴트 = 평균
```

```python
#상위 10개 정도 뽑을 때 : value_counts() 사용 -> 이거 사용해서 유입이 가장 많은 곳 확인
#문제 : 채널 유입이 높다고 구매율이 높은건 아님 -> 즉, value_counts()로 단순히 판단 불가
df1['channel'].value_counts()
```

```python
#pivot_table 활용
p1 = pd.pivot_table(data = df1, index = 'channel', values = 'amount', aggfunc = 'sum')
p1
# 합계산 : 유입채널에 따른 판매액
# 정렬 안되있음 => 정렬
# 매출액이 가장높은 상위 10개의 채널
p1.sort_values(by ='amount', ascending = False).head(10) #내림차순 ascending = False
```

```python
df1['channel'].value_counts()
#피봇테이블을 통해 확인한 결과와 R관리방판 이후로 달라짐
#=>(하이마트가 영업방판보다 방문채널은 크지만 매출액은 작음)
# 이런 결과를 눈여겨봐서 문제해결
```

```python
# 매출액이 가장높은 상위 10개의 채널 인덱스 추출
top_list = p1.sort_values(by ='amount', ascending = False).head(5).index.to_list()
top_list
```

```python
#상위 10개 데이터에 대해서 출력
plt.figure(figsize = [20, 10])
sns.barplot(data = df1, x = 'channel', y = 'amount', order = top_list)
#그래프를 보면 일단 매출액 가장 큰 상위 5개만 뽑아준건데 y값이 평균값이므로 그래프 보기에는 뭐가 1윈지 모름
#따라서 매출액의 평균이 아닌 합을 구해서 표현해야함
```

```python
plt.figure(figsize = [20, 10])
sns.barplot(data = df1, x = 'channel', y = 'amount', order = top_list, estimator = sum) #estimator = sum 
```

```python
#위의 그래프 보면 막대 중간에 검은색 점이 있는데 이건 신뢰구간값 표시된거
#지우기 위해서는 ci = None 사용
plt.figure(figsize = [20, 10])
sns.barplot(data = df1, x = 'channel', y = 'amount', order = top_list, estimator = sum, ci = None) #estimator = sum 
```

```python
#바차트 오버레이 확인
plt.figure(figsize = [20, 10])
sns.barplot(data = df1, x = 'channel', y = 'amount', order = top_list, estimator = sum, ci = None,
           hue = 'type_of_contract')
```

```python
# boxplot : 각각 범주형 항목에 대해 데이터의 분포를 상자형태로 출력
plt.figure(figsize = [20, 10])
sns.boxplot(data = df1, x = 'channel', y = 'amount', order = top_list)
#채널 별 매출액 분포가 어캐 되어있는지
```
