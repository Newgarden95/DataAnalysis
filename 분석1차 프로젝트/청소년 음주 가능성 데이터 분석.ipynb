### 데이터 전처리 과정 : 음주 가능성 
#### 주중 + 주말 음주소비 > 2 이면  : 1(음주를 할것이다) , 0(음주를 하지 않을 것이다.)
###### 라벨링 , for문(리스트에 0, 1 쌓기-> 조건문통해)

#### 1. for문 이용

```python
# 1번째 방법 for문 이용하기
drink = []
raw_data['drink_possibility'] = raw_data['Dalc'] + raw_data['Walc']
raw_data['drink_possibility']
dp = []
for i in raw_data['drink_possibility']:
    if(i == 2):
        dp.append(0)
    else:
        dp.append(1)
dp

raw_data['dp'] = dp
raw_data
```

#### 2. 라벨링 이용

```python
# 2번째 방법 라벨링하기
raw_data['drink_possibility'].unique()
```

```python
# 라벨링 하기
mapping = {2 : 0, 3 : 1, 4 : 1, 5 : 1, 6 : 1, 7 : 1, 8 : 1, 9 : 1, 10 : 1}
```

```python
#map함수를 통해 라벨링 하기
raw_data['drink_possibility'] = raw_data['drink_possibility'].map(mapping)
```

```python
#변경된 값 확인하기
raw_data
```

# 보호자-음주가능성

### 인사이트
#### 보호자는 엄마가 월등히 많다
#### 음주할 확률은 보호자가 아빠, 엄마, 기타등등 의 경우 거의 비슷하지만
#### 표본이 달라 정확히 비교는 불가능하다
#### 따라서, 보호자에 따른 학생의 음주는 연관이 없다

```python
#https://rfriend.tistory.com/383
#전체 데이터를 그룹 별로 나누고 (split), 각 그룹별로 집계함수를 적용(apply) 한후, 그룹별 집계 결과를 하나로 합치는(combine) 단계

g_d = raw_data[['guardian', 'dp']]
g_d

g_d.groupby('guardian').count() #보호자에 따른 음주여부 카운트
```

```python
# 보호자를 그룹으로 음주를 한 학생
g_yd = g_d.groupby('guardian').sum().reset_index()
g_yd
```

```python
# 보호자에 따라 음주를 하지 않는 학생
g_nd = g_d[g_d['dp'] == 0]
g_nd
 
g_nd = g_nd.groupby('guardian').count().reset_index() #.count() 
g_nd['guardian'].to_list()
```

```python
f, ax = plt.subplots(1,3, figsize = [12, 8])
ax[0].set_title("보호자에 따른 음주경험있는 학생의 수")
sns.barplot(data = g_yd, x = 'guardian', y = 'dp', ax = ax[0])

ax[1].set_title("보호자에 따른 음주경험없는 학생의 수")
sns.barplot(data = g_nd, x = 'guardian', y = 'dp', ax = ax[1])

ax[2].set_title("보호자에 따른 음주경험있는 학생의 비율")
g_d.groupby('guardian').sum().reset_index()['dp'].plot.pie(autopct = "%1.2f%%", ax = ax[2])
ax[2].set_xlabel("0 : father 1 : mother 2 :other")
```
